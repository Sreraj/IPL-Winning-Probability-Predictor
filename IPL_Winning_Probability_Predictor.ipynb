```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals',
    'Gujarat Titans',
    'Lucknow Super Giants'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df['venue'] = match_df['venue'].replace("M.Chinnaswamy Stadium","M Chinnaswamy Stadium,Bengaluru")
match_df['venue'] = match_df['venue'].replace("M Chinnaswamy Stadium","M Chinnaswamy Stadium,Bengaluru")
match_df['venue'] = match_df['venue'].replace("M Chinnaswamy Stadium, Bengaluru","M Chinnaswamy Stadium,Bengaluru")

match_df['venue'] = match_df['venue'].replace("MA Chidambaram Stadium, Chepauk","MA Chidambaram Stadium, Chepauk, Chennai")
match_df['venue'] = match_df['venue'].replace("MA Chidambaram Stadium","MA Chidambaram Stadium, Chepauk, Chennai")

match_df['venue'] = match_df['venue'].replace("Wankhede Stadium","Wankhede Stadium, Mumbai")
match_df['venue'] = match_df['venue'].replace("Eden Gardens","Eden Gardens, Kolkata")

match_df['venue'] =match_df['venue'].replace("Feroz Shah Kotla, Delhi","Arun Jaitley Stadium, Delhi")
match_df['venue'] = match_df['venue'].replace("Feroz Shah Kotla","Arun Jaitley Stadium, Delhi")
match_df['venue'] = match_df['venue'].replace("Arun Jaitley Stadium","Arun Jaitley Stadium, Delhi")

match_df['venue'] = match_df['venue'].replace("Punjab Cricket Association IS Bindra Stadium, Mohali, Chandigarh","Punjab Cricket Association Stadium, Mohali")
match_df['venue'] = match_df['venue'].replace("Punjab Cricket Association IS Bindra Stadium, Mohali","Punjab Cricket Association Stadium, Mohali")
match_df['venue'] =match_df['venue'].replace("Punjab Cricket Association IS Bindra Stadium","Punjab Cricket Association Stadium, Mohali")

match_df['venue'] = match_df['venue'].replace("Rajiv Gandhi International Stadium, Uppal","Rajiv Gandhi International Stadium")
match_df['venue'] = match_df['venue'].replace("Rajiv Gandhi International Stadium, Uppal, Hyderabad","Rajiv Gandhi International Stadium")
```

```python
match_df.shape
```

```python
match_df = match_df[['match_id','venue','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df
```

```python
# current_score
delivery_df['current_score'] = delivery_df.groupby(['match_id'])['total_runs_y'].cumsum()
# runs left
delivery_df['runs_left'] =delivery_df['total_runs_x'] - delivery_df['current_score']
# balls_left
delivery_df['balls_left'] = 120 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x : x if x=='0' else '1' )
delivery_df['player_dismissed']=delivery_df['player_dismissed'].astype(int)

wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum()

# wicket
delivery_df['wickets'] = 10 - wickets
# crr ====>    runs/over
delivery_df['crr'] =  delivery_df['current_score'] * 6/ (120- delivery_df['balls_left'])
# rrr ===>  runs / balls left
delivery_df['rrr'] = (delivery_df['runs_left'] * 6) /delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0

delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
final_df = delivery_df[['batting_team','bowling_team','venue','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df.isna().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
x = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
# Encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(handle_unknown='ignore',drop='first'),['batting_team','bowling_team','venue'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression


from sklearn.pipeline import Pipeline
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
